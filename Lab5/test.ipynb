Q1. Plot the pole zero locations and time domain impulse response (clearly mark the point n=0 ) for the following transfer functions.
<ol>
  <li>$ \frac{1}{1+az^{-1}} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
 <li>$ \frac{1}{(1+az^{-1})^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
  <li>$\frac{1}{1-az^{-1}+{(az^{-1})}^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Complex conjugate poles) </li>    
</ol>
Which of the responses corresponds to stable LTI systems?

https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.signal.tf2zpk.html

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
#***********************************

#1

#a = 0.8
b=[1]
a=[1,0.8]
z,p,k =sig.tf2zpk(b,a)
print(z,p)

plt.figure(figsize=(5, 5))
plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
n1 = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n1),np.cos(n1))
plt.show()

#a = -0.8
ave=[1,-0.8]
zve,pve,kve =sig.tf2zpk(b,ave)
print(zve,pve)

plt.figure(figsize=(5, 5))
plt.plot(zve.real, zve.imag, 'ro', pve.real, pve.imag, 'rx')
n1ve = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n1ve),np.cos(n1ve))
plt.show()

#a = 1
a1=[1,1]
z1,p1,k1 =sig.tf2zpk(b,a1)
print(z1,p1)

plt.figure(figsize=(5, 5))
plt.plot(z1.real, z1.imag, 'ro', p1.real, p1.imag, 'rx')
n2 = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n2),np.cos(n2))
plt.show()

#a = -1
a1ve=[1,-1]
z1ve,p1ve,k1ve =sig.tf2zpk(b,a1ve)
print(z1,p1)

plt.figure(figsize=(5, 5))
plt.plot(z1ve.real, z1ve.imag, 'ro', p1ve.real, p1ve.imag, 'rx')
n2ve = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n2ve),np.cos(n2ve))
plt.show()

#a = 2
a2=[1,2]
z2,p2,k2 =sig.tf2zpk(b,a2)
print(z2,p2)

plt.figure(figsize=(5, 5))
plt.plot(z2.real, z2.imag, 'ro', p2.real, p2.imag, 'rx')
n3 = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n3),np.cos(n3))
plt.show()

#a = -2
a2ve=[1,-2]
z2ve,p2ve,k2ve =sig.tf2zpk(b,a2ve)
print(z2ve,p2ve)

plt.figure(figsize=(5, 5))
plt.plot(z2ve.real, z2ve.imag, 'ro', p2ve.real, p2ve.imag, 'rx')
n3ve = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n3ve),np.cos(n3ve))
plt.show()
```

  <li>$ 1 + 2z^{-1} + 3z^{-2} + 2z^{-3} + z^{-4}$ </li>

   <li>$ 1 + 2z^{-1} + 3z^{-2} + 4z^{-3} + 5z^{-4} $ </li>

Where are the poles located? What is the type of response in each case? What is the advantage of symmetric/anti-symmetric filters?

Q3. Compute the causal inverse of $H(z) = \frac{z^{-1}+0.5z^{-2}}{1-0.6z^{-1}+2z^{-2}}$. (use <i>scipy.signal.residuez</i>) Is the system stable?

https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.residuez.html

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

b = [0.0, 1.0, 0.5]
a = [1.0, -0.6, 2.0]
r, p, k = sig.residuez(b, a, tol=0.001, rtype='avg')

z,p,k =sig.tf2zpk(b,a)
print(z,p)

plt.figure(figsize=(5, 5))
plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
n1 = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n1),np.cos(n1))
plt.show()
```
