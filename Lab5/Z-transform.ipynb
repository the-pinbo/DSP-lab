## Z-Transform

Q1. Plot the pole zero locations and time domain impulse response (clearly mark the point n=0 ) for the following transfer functions.
<ol>
  <li>$ \frac{1}{1+az^{-1}} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
 <li>$ \frac{1}{(1+az^{-1})^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
  <li>$\frac{1}{1-az^{-1}+{(az^{-1})}^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Complex conjugate poles) </li>    
</ol>
Which of the responses corresponds to stable LTI systems?

https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.signal.tf2zpk.html

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

<li>$ \frac{1}{1+az^{-1}} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>

```python
a = [.8, 1, 2]
for i in range(3):
    for j in range(2):
        figure, axis = plt.subplots(1, 2, figsize=(15, 5))
        a0 = a[i] if j == 0 else -a[i]
        _b = [1]
        _a = [1, a0]
        z, p, k = sig.tf2zpk(_b, _a)
        figure.suptitle(f'For a = {a0}')
        axis[0].plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
        axis[0].set_title(f"Pole/Zero plot:Zeros = {z}, Poles = {p}")
        n1 = np.arange(0, np.pi * 2, 1 / 100)
        axis[0].plot(np.cos(n1), np.sin(n1))
        x = sig.dlti(_b, _a)
        t, y = sig.dimpulse(x, n=25)
        axis[1].stem(t, np.squeeze(y))
        axis[0].grid()
        axis[1].grid()
        axis[1].set_title(f"Time domain impulse reponse a = {a0}")
        plt.show()
```

 <li>$ \frac{1}{(1+az^{-1})^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>


```python
a = [.8, 1, 2]
for i in range(3):
    for j in range(2):
        a_c = a[i] if j == 0 else -a[i]
        _b = [1]
        _a = [1, 2 * a_c, a_c**2]
        z, p, k = sig.tf2zpk(_b, _a)
        figure.suptitle(f'For a = {a_c}')
        axis[0].plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
        axis[0].set_title(f"Pole/Zero plot:Zeros = {z}, Poles = {p}")
        n1 = np.arange(0, np.pi * 2, 1 / 100)
        axis[0].plot(np.cos(n1), np.sin(n1))
        x = sig.dlti(_b, _a)
        t, y = sig.dimpulse(x, n=25)
        axis[1].stem(t, np.squeeze(y))
        axis[0].grid()
        axis[1].grid()
        axis[1].set_title(f"Time domain impulse reponse a = {a_c}")
        plt.show()
```

  <li>$\frac{1}{1-az^{-1}+{(az^{-1})}^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Complex conjugate poles) </li> 

```python
a = [.8, 1, 2]
for i in range(3):
    for j in range(2):
        a_c = a[i] if j == 0 else -a[i]
        _b = [1]
        _a = [1, -a_c, a_c**2]
        z, p, k = sig.tf2zpk(_b, _a)
        figure.suptitle(f'For a = {a_c}')
        axis[0].plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
        axis[0].set_title(f"Pole/Zero plot:Zeros = {z}, Poles = {p}")
        n1 = np.arange(0, np.pi * 2, 1 / 100)
        axis[0].plot(np.cos(n1), np.sin(n1))
        x = sig.dlti(_b, _a)
        t, y = sig.dimpulse(x, n=25)
        axis[1].stem(t, np.squeeze(y))
        axis[0].grid()
        axis[1].grid()
        axis[1].set_title(f"Time domain impulse reponse a = {a_c}")
        plt.show()
```
