```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig


def polezero(b,a):

  z,p,k =sig.tf2zpk(b,a)
  print(z,p)

  plt.figure(figsize=(5, 5))
  plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
  n1 = np.arange(0,np.pi*2,1/100)
  plt.plot(np.sin(n1),np.cos(n1))
  plt.show()

#1 part a
polezero([1,0],[1,0.8])
#1 part b
polezero([1,0],[1,-0.8])
#1 part c
polezero([1,0],[1,1])
#1 part d
polezero([1,0],[1,-1])
#1 part e
polezero([1,0],[1,2])
#1 part f
polezero([1,0],[1,-2])




```

Stability:

1.A discrete LTI system is stable if and only if ROC of its transfer function includes the unit circle |z| = 1.

2.A causal discrete LTI system with rational transfer function H(z) is stable if and only if all the poles of H(z) lie inside the unit circle |z|= 1.

3.So , we need to check the poles location of the z tranforms given to us.

4.For the first part of question 1, we see for a=0.8,-0.8 ; the system is stable as the poles lie within the unit circle.

5.For a=1,-1 ; the system can be said to be marginally stable as the poles lie on the unit circle.

6.For a=2,-2 ;the system will be unstable because the poles lie outside the unit circle.

Q2. Plot the pole zero locations of the following finite impulse response functions. 
<ol>
  <li>$ z + 2 + z^{-1}$ </li>
  <li>$ 1 + 2z^{-1} + 3z^{-2} + 2z^{-3} + z^{-4}$ </li>
  <li>$ 1 + 2z^{-1} - 2z^{-3} - z^{-4} $ </li>
   <li>$ 1 + 2z^{-1} + 3z^{-2} + 4z^{-3} + 5z^{-4} $ </li>    
</ol>
Where are the poles located? What is the type of response in each case? What is the advantage of symmetric/anti-symmetric filters?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig


def polezero(b,a):

  z,p,k =sig.tf2zpk(b,a)
  print(z,p)

  plt.figure(figsize=(5, 5))
  plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
  n1 = np.arange(0,np.pi*2,1/100)
  plt.plot(np.sin(n1),np.cos(n1))
  plt.show()

#part 1
polezero([1,2,1],[1,0])
#part 2
polezero([1,2,3,2,1],[1,0])
#part 3
polezero([1,2,0,-2,-1],[1,0])
#part 4
polezero([1,2,3,4,5],[1,0])

```

The poles of a transfer function is the value of z for which it becomes infinite or undefined. The poles of all the systems given in this question are located at origin.

The first one will be a type 1 system, as it has one pole at origin.
The second one will be a type 4 system, as it has four poles at origin.
The third one will be a type 4 system, as it has four poles at origin.
The fourth one will be a type 4 system, as it has four poles at origin.

Since for all the systems the poles are located inside the unit circle, we can say they correspond to stable systems.

Advantages of symmetric and antisymmetric filters is we exploit the symmetricity in the coefficients of LTI system, which eases the calculations and reduces the complexity of the filter. Moreover, coefficient symmetricity produces linear phase response.

with zero initial conditions. ( Use scipy.sig.lfilter )

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig


n=np.arange(-1,25)
#unit step signal
def unit_step(t,n1):
  step_signal=[]
  for a in n1:
    if(a>=t):
      step_signal.append(1)
    elif(a<t):
      step_signal.append(0)
  return step_signal


#fxn1:
b=[0.5,0.5]
a=[1.0,-0.8]

step1=unit_step(0,n)
step2=unit_step(10,n)
#for initial condition y[-1]=-2
input1=[x1-x2 for (x1,x2) in zip(step1,step2)]
input2=[2*x1 for x1 in input1]
initial_condition=sig.lfiltic(b,a,y=[-2.0/0.8])  #initial condition is set like this to make sure y[-1]=-2 , which can be seen in the plots
response1,response=sig.lfilter(b,a,input1,zi=initial_condition) 
response1=list(response1) 
response=list(response)
response2,res=sig.lfilter(b,a,input2,zi=initial_condition) 
response2=list(response2) 
res=list(res)
plt.stem(n,response1,use_line_collection=True)
plt.xticks(np.arange(min(n), max(n)+1, 1.0))
plt.show()
plt.stem(n,response2,use_line_collection=True)
plt.xticks(np.arange(min(n), max(n)+1, 1.0))
plt.show()


#for initial condition y[-1]=0
input3=[x1-x2 for (x1,x2) in zip(step1,step2)]
input4=[2*x1 for x1 in input3]
response3=sig.lfilter(b,a,input3) 
response3=list(response3) 
response4=sig.lfilter(b,a,input4) 
response4=list(response4) 

plt.stem(n,response3,use_line_collection=True)
plt.xticks(np.arange(min(n), max(n)+1, 1.0))
plt.show()
plt.stem(n,response4,use_line_collection=True)
plt.xticks(np.arange(min(n), max(n)+1, 1.0))
plt.show()


```

For the second part of the question , they have asked us to assume zero initial conditions. In that case , the system would be linear.

Q5. For the LTI systems described by the following difference equation. Obtain the pole zero plot. Comment on the type of response.

$y[n] + 1.3y[n - 1] + 1.29y[n - 2] + 0.52y[n - 3] + 0.16y[n - 4] = x[n] - 0.9x[n - 1]$

$y[n] - 1.3y[n - 1] + 1.29y[n - 2] - 0.52y[n - 3] + 0.16y[n - 4] = x[n] + 0.9x[n - 1]$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig


def polezeroq4(b,a):

  z,p,k =sig.tf2zpk(b,a)
  print(z,p)

  plt.figure(figsize=(5, 5))
  plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
  n1 = np.arange(0,np.pi*2,1/100)
  plt.plot(np.sin(n1),np.cos(n1))
  plt.show()

#4 part a
polezeroq4([1,-0.9,0],[1,1.3,1.29,0.52,0.16])
#4 part b
polezeroq4([1,0.9,0],[1,-1.3,1.29,-0.52,0.16])



```

Type of response:

We can see the system in time domain is causal in nature(being deoendent on the present and past values of n). 

A causal discrete LTI system with rational transfer function H(z) is stable if and only if all the poles of H(z) lie inside the unit circle |z|= 1.

We can see that for both the cases , the poles lie inside the unit circle. Hence , the responses correspond to stable LTI system.

Also , based on the number of poles , we can comment on the type and order of system. In both the cases , we can see , the system has 4 poles. The type of the system is the number of poles at the origin. Since, here there are no poles at the origin , we can say that the system is type 0. The order of the system is 4 (maximum power of z in denominator).
