## Z-Transform

Q1. Plot the pole zero locations and time domain impulse response (clearly mark the point n=0 ) for the following transfer functions.
<ol>
  <li>$ \frac{1}{1+az^{-1}} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
 <li>$ \frac{1}{(1+az^{-1})^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
  <li>$\frac{1}{1-az^{-1}+{(az^{-1})}^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Complex conjugate poles) </li>    
</ol>
Which of the responses corresponds to stable LTI systems?

https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.signal.tf2zpk.html

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

<li>$ \frac{1}{1+az^{-1}} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>

```python
a = [.8, 1, 2]
for i in range(3):
    for j in range(2):
        figure, axis = plt.subplots(1, 2, figsize=(15, 5))
        a0 = a[i] if j == 0 else -a[i]
        _b = [1, 0]
        _a = [1, a0]
        z, p, k = sig.tf2zpk(_b, _a)
        figure.suptitle(f'For a = {a0}')
        axis[0].plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
        axis[0].set_title(f"Pole/Zero plot:Zeros = {z}, Poles = {p}")
        n1 = np.arange(0, np.pi * 2, 1 / 100)
        axis[0].plot(np.cos(n1), np.sin(n1))
        x = sig.dlti(_b, _a)
        t, y = sig.dimpulse(x, n=25)
        axis[1].stem(t, np.squeeze(y))
        axis[0].grid()
        axis[1].grid()
        axis[1].set_title(f"Time domain impulse reponse a = {a0}")
        plt.plot(0,
                 0,
                 marker="o",
                 markersize=10,
                 markeredgecolor="red",
                 markerfacecolor="green"
                 )  #to mark n=0 clearly as said in the question
        plt.show()
```

### A causal discrete LTI system with rational transfer function H(z) is stable if and only if all the poles of H(z) lie inside the unit circle |z|= 1.

a.For a=0.8,-0.8 ; the system is stable as the poles lie within the unit circle.

b.For a=1,-1 ; the system can be said to be marginally stable as the poles lie on the unit circle.

c.For a=2,-2 ;the system will be unstable because the poles lie outside the unit circle.

 <li>$ \frac{1}{(1+az^{-1})^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>


```python
a = [.8, 1, 2]
for i in range(3):
    for j in range(2):
        figure, axis = plt.subplots(1, 2, figsize=(15, 5))
        a_c = a[i] if j == 0 else -a[i]
        _b = [1, 0, 0]
        _a = [1, 2 * a_c, a_c**2]
        z, p, k = sig.tf2zpk(_b, _a)
        figure.suptitle(f'For a = {a_c}')
        axis[0].plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
        axis[0].set_title(f"Pole/Zero plot:Zeros = {z}, Poles = {p}")
        n1 = np.arange(0, np.pi * 2, 1 / 100)
        axis[0].plot(np.cos(n1), np.sin(n1))
        x = sig.dlti(_b, _a)
        t, y = sig.dimpulse(x, n=25)
        axis[1].stem(t, np.squeeze(y))
        axis[0].grid()
        axis[1].grid()
        axis[1].set_title(f"Time domain impulse reponse a = {a_c}")
        plt.plot(0,
                 0,
                 marker="o",
                 markersize=10,
                 markeredgecolor="red",
                 markerfacecolor="green"
                 )  #to mark n=0 clearly as said in the question
        plt.show()
```

1.For a=0.8,-0.8 ; the system is stable 

2.For a=1,-1 ; the system can be said to be marginally stable

3.For a=2,-2 ;the system will be unstable

  <li>$\frac{1}{1-az^{-1}+{(az^{-1})}^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Complex conjugate poles) </li> 

```python
a = [.8, 1, 2]
for i in range(3):
    for j in range(2):
        figure, axis = plt.subplots(1, 2, figsize=(15, 5))
        a_c = a[i] if j == 0 else -a[i]
        _b = [1, 0, 0]
        _a = [1, -a_c, a_c**2]
        z, p, k = sig.tf2zpk(_b, _a)
        figure.suptitle(f'For a = {a_c}')
        axis[0].plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
        axis[0].set_title(f"Pole/Zero plot:Zeros = {z}, Poles = {p}")
        n1 = np.arange(0, np.pi * 2, 1 / 100)
        axis[0].plot(np.cos(n1), np.sin(n1))
        x = sig.dlti(_b, _a)
        t, y = sig.dimpulse(x, n=25)
        axis[1].stem(t, np.squeeze(y))
        axis[0].grid()
        axis[1].grid()
        axis[1].set_title(f"Time domain impulse reponse a = {a_c}")
        plt.plot(0,
                 0,
                 marker="o",
                 markersize=10,
                 markeredgecolor="red",
                 markerfacecolor="green"
                 )  #to mark n=0 clearly as said in the question
        plt.show()
```

1.For a=0.8,-0.8 ; the system is stable

2.For a=1,-1 ; the system can be said to be marginally stable

3.For a=2,-2 ;the system will be unstable

Q3. Compute the causal inverse of $H(z) = \frac{z^{-1}+0.5z^{-2}}{1-0.6z^{-1}+2z^{-2}}$. (use <i>scipy.signal.residuez</i>) Is the system stable?

https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.residuez.html

```python
b = [1, 0.5]
a = [1, -0.6, 2]

z, p, k = sig.tf2zpk(b, a)
print(z, p)

plt.figure(figsize=(5, 5))
plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
r = abs(p[0])
plt.title(f"Pole/Zero plot\nZeros = {z}, Poles = {p}\n r ={r}")
n1 = np.arange(0, np.pi * 2, 1 / 100)
plt.plot(np.cos(n1), np.sin(n1))

n2 = np.arange(0, np.pi * 2, 1 / 10)
plt.plot(r * np.cos(n2), r * np.sin(n2), 'r.')
plt.grid()
plt.show()
```

The system is unstable because, |z| = 1 is not included in the ROC, as there is a pole at r = 1.141.
ROC |z| > 1.14142

```python
n = np.arange(-5, 20)


def unit_step(t, n1):
    step_signal = []
    for a in n1:
        if (a >= t):
            step_signal.append(1)
        elif (a < t):
            step_signal.append(0)
    return step_signal


def unit_impulse(a, n):
    delta = []
    for sample in n:
        if sample == a:
            delta.append(1)
        else:
            delta.append(0)

    return delta
```

```python
delta = unit_impulse(0, n)
unit_fun = unit_step(0, n)
b = [0, 1, 0.5]
a = [1, -0.6, 2]

r, p, k = sig.residuez(b, a)
delta = [k[0] * x1 for x1 in delta]
result = r[0] * (p[0]**n) + r[1] * (p[1]**n) + delta
result = [x1 * x2 for (x1, x2) in zip(result, unit_fun)]

plt.stem(n, result, use_line_collection=True)
plt.title("Causal Inverse")
plt.show()
```

We are using scipy.signal.residuez function to decompose the given transfer function into its partial fractions. We obtain three matrices: r, p and k. We find that the r and p matrices for the transfer function given to us are complex in nature. 

So , firstly we use residuez function to get the partial fractions , then use the formula we have learnt in dsp theory to compute the inverse of the transform given.

Q5. For the LTI systems described by the following difference equation. Obtain the pole zero plot. Comment on the type of response.

$y[n] + 1.3y[n - 1] + 1.29y[n - 2] + 0.52y[n - 3] + 0.16y[n - 4] = x[n] - 0.9x[n - 1]$



```python
a = [1, 1.3, 1.29, 0.52, 0.16]
b = [1, -0.9, 0]

z, p, k = sig.tf2zpk(b, a)
print(z, p)

plt.figure(figsize=(5, 5))
plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')

plt.title(f"Pole/Zero plot\nZeros = {z}, Poles = {p}")
n1 = np.arange(0, np.pi * 2, 1 / 100)
plt.plot(np.cos(n1), np.sin(n1))

plt.grid()
plt.show()
```

As the poles lie inside the unit circle. Hence stable LTI system.

$y[n] - 1.3y[n - 1] + 1.29y[n - 2] - 0.52y[n - 3] + 0.16y[n - 4] = x[n] + 0.9x[n - 1]$

```python
a = [1, -1.3, 1.29, -0.52, 0.16]
b = [1, 0.9, 0]

z, p, k = sig.tf2zpk(b, a)
print(z, p)

plt.figure(figsize=(5, 5))
plt.plot(z.real, z.imag, 'go', p.real, p.imag, 'rx')

plt.title(f"Pole/Zero plot\nZeros = {z}, Poles = {p}")
n1 = np.arange(0, np.pi * 2, 1 / 100)
plt.plot(np.cos(n1), np.sin(n1))

plt.grid()
plt.show()
```

As the poles lie inside the unit circle. Hence stable LTI system.

Type of response:

We can see the system in time domain is causal in nature(being deoendent on the present and past values of n). 

A causal discrete LTI system with rational transfer function H(z) is stable if and only if all the poles of H(z) lie inside the unit circle |z|= 1.

We can see that for both the cases , the poles lie inside the unit circle. Hence , the responses correspond to stable LTI system.

Also , based on the number of poles , we can comment on the type and order of system. In both the cases , we can see , the system has 4 poles. The type of the system is the number of poles at the origin. Since, here there are no poles at the origin , we can say that the system is type 0. The order of the system is 4 (maximum power of z in denominator).
