### CODE

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
N = 4096
omega = 2 * np.pi
n = 15
x = np.arange(-n, n + 1)
x
```

1) $r[n] = u[n] − u[n − 5]$

```python
n0 = 0
```

```python
r = np.zeros(x.size)
r[n + n0:n + n0 + 5] = 1
plt.stem(x, r)
```

```python
def plot(r):
    a = np.ones(1)
    b = r[n:]
    c = c = np.arange(-omega, omega, 2 * omega / N)
    w1, h1 = sig.freqz(b, a, c)
    #     magnitude
    h1db = 20 * np.log10(abs(h1))
    plt.plot(w1 / omega, h1db)
    plt.show()
    #     phase
    angles = np.angle(h1, deg=True)
    plt.plot(w1 / omega, angles, 'g')
    plt.show()
```

```python
plot(r)
```

2) $r[n − 7]$ 

```python
n0 = 7
```

```python
r = np.zeros(x.size)
r[n + n0:n + n0 + 5] = 1
plt.stem(x, r)
```

```python
plot(r)
```

3) $r[n + 4]$ 

```python
n0 = -4
```

```python
r = np.zeros(x.size)
r[n + n0:n + n0 + 5] = 1
plt.stem(x, r)
```

```python
plot(r)
```

4) $r[−n]$

```python
n0 = 0
```

```python
r = np.zeros(x.size)
r[n + n0:n + n0 + 5] = 1
x = -x
plt.stem(x, r)
```

```python
plot(r)
```

5) $(−1)^nr[n]$ 

```python
n0 = 0
```

```python
r = np.zeros(x.size)
r[n + n0:n + n0 + 5] = 1
r[x%2 == 1] *= -1
plt.stem(x, r)
```

```python
plot(r)
```
